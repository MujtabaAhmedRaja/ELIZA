```python
def is_variable(pattern):
    return (type(pattern) is str
            and pattern[0] == '?'
            and len(pattern) > 1
            and pattern[1] != '*'
            and pattern[1] in letters
            and ' ' not in pattern)
```

```python
def contains_tokens(pattern):
    return type(pattern) is list and len(pattern) > 0
```

```python
def match_variable(var, replacement, bindings):
    binding = bindings.get(var)
    if not binding:
        bindings.update({var: replacement})
        return bindings
    if replacement == bindings[var]:
        return bindings
    return False
```

```python
def is_segment(pattern):
    return (type(pattern) is list
            and pattern
            and pattern and len(pattern[0]) > 2
            and pattern[0][0] == '?'
            and pattern[0][1] == '*'
            and pattern[0][2] in letters
            and ' ' not in pattern[0])
```

```python
def match_segment(var, pattern, input, bindings):
    if not pattern:
        return match_variable(var, input, bindings)
    word = pattern[0]
    try:
        pos = input.index(word)
    except ValueError:
        return False
    var_match = match_variable(var, input[:pos], dict(bindings))
    match = match_pattern(pattern, input[pos:], var_match)
    # if not match:
    #     return match_segment(var, pattern, input, bindings)
    return match
```

```python
def match_pattern(pattern, input):
        """
        Matches a pattern with an input string and returns a dictionary of bindings.
        Pattern variables start with '?'.
        """
        bindings = {}
        if len(pattern) > len(input) and not any(p.startswith("?*") for p in pattern):
            return None  # Pattern longer than input and no wildcards
    
        i, j = 0, 0
        while i < len(pattern) and j < len(input):
            if pattern[i].startswith("?*"):
                var = pattern[i][2:]  # Extract the variable name (e.g., X from ?*X)
                remaining_pattern = pattern[i + 1:]
                remaining_input = input[j:]
                
                for k in range(len(remaining_input) + 1):
                    bindings[var] = remaining_input[:k]
                    result = match_pattern(remaining_pattern, remaining_input[k:])
                    if result is not None:
                        bindings.update(result)
                        return bindings
                return None
            elif pattern[i].startswith("?"):
                var = pattern[i][1:]  # Extract the variable name (e.g., X from ?X)
                bindings[var] = [input[j]]
                i += 1
                j += 1
            elif pattern[i] == input[j]:
                i += 1
                j += 1
            else:
                return None
        
        if i < len(pattern) and not pattern[i].startswith("?*"):
            return None  # Pattern not fully matched
        return bindings
    
# def match_pattern(pattern, input, bindings=None):
#     if bindings is False:
#         return False
#     if pattern == input:
#         return bindings
#     bindings = bindings or {}
#     if is_segment(pattern):
#         token = pattern[0]
#         var = token[2:]
#         return match_segment(var, pattern[1:], input, bindings)
#     elif is_variable(pattern):
#         var = pattern[1:]
#         return match_variable(var, [input], bindings)
#     elif contains_tokens(pattern) and contains_tokens(input):
#         return match_pattern(pattern[1:],input[1:],match_pattern(pattern[0], input[0], bindings))
#     else:
#         return False
#######################################################################
```
